```python
import pandas as pd

data = pd.read_csv('airplane_crashes.csv')
data
```

```python
missing_values = data.isnull().sum()

# Displaying missing values
missing_values
```

```python

data_cleaned = data.dropna()

features = data_cleaned[['Fatalities', 'Ground', 'Aboard']]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

features_scaled_df = pd.DataFrame(features_scaled, columns=features.columns)

# Displaying the normalized features
features_scaled_df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow method to find the optimal number of clusters
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled_df)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Applying KMeans with the chosen number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)
data_cleaned['Cluster'] = kmeans.fit_predict(features_scaled_df)

# Displaying the first few rows with cluster labels
data_cleaned
```

```python
import seaborn as sns

# Visualizing the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Aboard', y='Fatalities', hue='Cluster', data=data_cleaned, palette='viridis')
plt.title('Clusters of Airplane Crashes')
plt.xlabel('Number of People Aboard')
plt.ylabel('Number of Fatalities')
plt.show()
```
